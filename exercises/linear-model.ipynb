# Linear model

We will analyze animal life-history data from [AnAge](http://genomics.senescence.info/download.html#anage). 
Specifically, we will try to find to examine [Kleiber's law](https://en.wikipedia.org/wiki/Kleiber%27s_law) which suggests that the metabolic rate of animals is proportional to their body mass raised to power 3/4, that is:
$$
W \sim m^{3/4}
$$
where $W$ is metabolic rate and $m$ is body mass.

Tasks:

1. Download and load the data from the URL below (use `pd.read_table` function instead of `pd.read_csv`).
1. Filter out rows with NaN body mass or body mass over 300 kg
1. Create a new column for the body mass raised to power 3/4
1. Calculate the linear model of metabolic rate as function of body mass 
1. Plot a scatterplot of metabolic rate vs body mass and add a line for the linear model

Bonus points:
1. Calculate a separate linear model for each class (taxonomy class, not OOP): mammals, reptiles, amphibians, and birds. Try to use [`df.groupby`](http://pandas.pydata.org/pandas-docs/stable/groupby.html) and iterate over the resulting data grouped dataframe.
1. Plot a separate scatterplot + linear model plot (you can use seaborn) for the different animal classes. Try to use the `hue` and/or `col` arguments of [`sns.lmplot`](http://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot).
